# Merge

```python
# Syntex
# pd.merge(left_df, right_df, how = 'inner', on=None, left_on=None, right_on=None)
```

```python
df1 = pd.DataFrame( {
    'ID': [1,2,3,4],
    'Name': ['Alice', 'Bob', 'Charlie', 'Mohit']
})

df2 = pd.DataFrame( {
    'ID': [1,2,3,5],
    'Score': [80, 70, 75, 100]
})
```

```python
df1
```

```python
df2
```

```python
# Inner Join
merge_df = pd.merge(df1, df2, how='inner', on='ID')
merge_df
```

```python
# Left Join
merge_df = pd.merge(df1, df2, how='left', on='ID')
merge_df
```

```python
# Right Join
merge_df = pd.merge(df1, df2, how='right', on='ID')
merge_df
```

```python
# Outer Join
merge_df = pd.merge(df1, df2, how='outer', on='ID')
merge_df
```

```python
# Merge with multiple keys
df1 = pd.DataFrame({
    'ID': [1,2,3],
    'Subject': ['Math', 'Science', 'English'],
    'Score': [100, 70, 88]
})

df2 = pd.DataFrame({
    'ID': [1,2,3],
    'Subject': ['Math', 'Science', 'English'],
    'Grade': ['A', 'B', 'A']
})

merge_df = pd.merge(df1, df2, how='inner', on=['ID', 'Subject'])
merge_df
```

```python
# Concatenating DataFrames
```

```python
# concat()
# pd.concat(objs, axis=0, join='outer', keys=None, ignore_index=False)
```

```python
df1 = pd.DataFrame( {
    'ID': [1,2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame( {
    'ID': [3,4],
    'Name': ['Mohit', 'Karan']
})

df3= pd.DataFrame( {
    'ID': [5,6],
    'Name': ['Charli', 'Arjun']
})
# axis=0 (row wise)
result = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
result
```

```python
# axis=1 (column wise)
result = pd.concat([df1, df2, df3], axis=1, keys = ['df1', 'df2', 'df3'])
result
```

# Group By In Pandas

Syntex
---
DataFrame.groupBy(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True)

```python
data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'East', 'North'],
    'Product': ['A'     , 'A',    'B',     'A',    'B',      'B',    'A'],
    'Sales':  [ 100,      200,     150,     100,     250,     300,    400],
    'Profit':  [10,       20,     15,        5,      25,       30,     50]
}
df = pd.DataFrame(data)
df
```

```python
group_region = df.groupby('Region')
group_region.loc['East', 'Region']
```

```python
# Group on Region to calculate total sales
for x, y in df.groupby('Region'):
    print(x)
    print(y)
    print("--------")
```

```python
grouped_sales = df.groupby('Region')['Sales'].sum()
grouped_sales
```

```python
# Group on Region to calculate total sales or each product
grouped_sales = df.groupby(['Region', 'Product'])['Sales'].sum()
grouped_sales
```

```python
# How we can add multiple aggregation
grouped_sales = df.groupby('Region')['Sales'].agg(['sum', 'mean', 'max'])
grouped_sales
```

```python
# Rename agg name
grouped_sales = df.groupby('Region')['Sales'].agg(total_sales='sum', avg_sales='mean', top_price='max')
grouped_sales
```

```python
# Transformation
# Add a column showing normalized sales per group
df['Noramalized_Sales'] = df.groupby('Region')['Sales'].transform(lambda x: x/x.sum())
df
```

```python
del df['Noramalized_Sales']
```

```python
# Filter
# Show all region sales where total sales > 400

filter_df = df.groupby('Region').filter(lambda x: x['Sales'].sum() > 400)
filter_df
```

```python
grouped_sales = df.groupby('Region')['Sales'].sum()
grouped_sales
# It's by default a Series and Region is consider as Index
```

```python
grouped_sales = df.groupby('Region', as_index=False)['Sales'].sum()
grouped_sales
```

```python
# Grouping using custom function
# Group by where the Region name start with 'N'
df.groupby(lambda x: df.loc[x, 'Region'].startswith('N'))['Sales'].sum()
```

```python
data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'East', 'North'],
    'Product': ['A'     , 'A',    'B',     'A',    'B',      'B',    'A'],
    'Sales':  [ 100,      200,     150,     100,     250,     300,    400],
    'Profit':  [10,       20,     15,        5,      25,       30,     50]
}
df = pd.DataFrame(data)
df
```

```python
pivot = df.pivot_table(values='Sales', index='Region', aggfunc='sum')
pivot
```

```python
pivot = df.pivot_table(values='Sales', index='Region', aggfunc=['sum', 'mean'])
pivot
```

```python
pivot = df.pivot_table(values='Sales', index='Region', columns='Product', aggfunc='sum')
pivot
```

```python
df = pd.read_csv('employee_performance.csv')
df
```

```python
# Task 1: Average By Department & Make it as a Index of df

pivot_avg = df.pivot_table(values=['Q1', 'Q2', 'Q3', 'Q4'], index='Department', aggfunc='mean')
pivot_avg
```

```python
# Task 2: Add total
pivot_avg_total = df.pivot_table(values=['Q1', 'Q2', 'Q3', 'Q4'], index='Department', aggfunc='sum', margins=True)
pivot_avg_total
```

```python
# Task 3: Calculate the max score for each department across all quaters
pivot_max = df.pivot_table(values=['Q1', 'Q2', 'Q3', 'Q4'], index='Department', aggfunc='max')
pivot_max
```

```python
# Task 4: Unpivot the quartely data into single column
melted = pd.melt(df, id_vars=['Employee_ID', 'Name', 'Department'], value_vars=['Q1', 'Q2', 'Q3', 'Q4'], var_name='Quater', value_name='Performance')
melted
```

```python
# Task 5: Group the melted data with Quater by and calculate the avg

avg_performance = melted.groupby('Quater')['Performance'].mean()
avg_performance
```
